```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
#Your df.info() shows:
#Order Date  object
#Ship Date   object
#This means they are stored as text, not dates.
#We must convert them.
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df[['Order Date','Year','Month']].head()
```

```python
# view data
df.head()
df.tail()

# check info
df.info()

# check missing values
df.isnull().sum()
```

```python
#This will show which category generates the most revenue.
#Categories in the dataset:
#Furniture
#Technology
#Office Supplies

sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

```python
#Regions usually are:
#West
#East
#Central
#South
#This helps analyze regional performance

sns.barplot(x='Region', y='Profit', data=df)
plt.title("Profit by Region")
plt.show()
```

```python
#This helps identify seasonal demand
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Best Selling Products")
plt.show()
```

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```
